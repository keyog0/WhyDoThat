```python
dtype_map = {
    'user_id' : 'string',
    'recruit_id' :'string',
    'datetime' : 'datetime'
}

def arr2str(array) :
    if array is not None :
        return ','.join(array)
    else :
        return None
def create_sql_item(item,dtype) :
        if dtype == 'string' or 'datetime':
            return "'{}'".format(item)
        elif dtype == 'bool' or 'int':
            return "{}".format(item)
def dict_list2string(items,dtype_map) :
    key_arr = []
    item_arr = []
    for key, item in items.items() :
        if key != 'id' and item != None :
            key_arr.append(key)
            item_arr.append(create_sql_item(item,dtype=dtype_map[key]))
    return arr2str(key_arr),arr2str(item_arr)
```

```python
import random

def input_random_data(user,recruit) :
    for rid in random.choices(recruit,k=100) :
        item = f"'{user}','{rid}','{str(datetime.now())}'"
        dc_db.insert_data('click','user_id,recruit_id,datetime',item)
    for rid in random.choices(recruit,k=100) :
        item = f"'{user}','{rid}','{str(datetime.now())}'"
        dc_db.insert_data('bookmark','user_id,recruit_id,datetime',item)
    for rid in random.choices(recruit,k=100) :
        item = f"'{user}','{rid}','{str(datetime.now())}'"
        dc_db.insert_data('recruit_apply','user_id,recruit_id,datetime',item)
    
```
